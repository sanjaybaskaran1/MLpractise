```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
X,y=make_blobs(n_samples=1000,centers=4,n_features=2)
```

```python
X
```

```python
y
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
## standardization--feature scaling technique
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

## lets build k-means model

```python
from sklearn.cluster import KMeans
```

```python
## we already know in k-means we use 
## wcss - within cluster sum of squares
## and elbow method identify suitable k-value
```

```python
## Elbow method To select K Value
wcss=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k,init="k-means++")
    kmeans.fit(X_train_scaled)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
k = [k for k in range(1,11)]
k
```

```python
plt.plot(k,wcss)
plt.xticks(range(1,11))
plt.xlabel("Number of Clustrers or k-value")
plt.title("elbow method")
plt.ylabel("WCSS")
plt.show()
```

## in the above curve the last decrease in the curve happen when k=4
## after that there is no change in the curve 
## so the suitable k-value is k=4
## which means we need four clusters to seperate our data

## so lets train our model with n_clusters=4

```python
kmeans=KMeans(n_clusters=4,init="k-means++")
```

```python
kmeans.fit_predict(X_train_scaled)
```

```python
y_pred=kmeans.predict(X_test_scaled)
```

```python
y_pred
```

```python
plt.scatter(X_test[:,0],X_test[:,1],c=y_pred)
```

## see in our prediction also we can able to get 4 clusters

## give your k and wcss in kneelocator it

```python
kl=KneeLocator(k,wcss,curve="convex",direction="decreasing")
```

```python
kl.elbow
```

## our elbow method says 3-clusters needed to seperate this data as clusters
## but our knee locater says just 3 clusters is enough
## always go with knee locator decision

## ---------------------------------------------------------------------------------------------------

## now lets see what our silhoute score tells about no.of.clusters

```python
## silhoutte score
from sklearn.metrics import silhouette_score
```

```python
## in silhoute k-value always starts with 2 not 1
```

```python
silhouette_coefficients=[]
for k in range(2,11):
    kmeans=KMeans(n_clusters=k,init="k-means++")
    kmeans.fit(X_train_scaled)
    score=silhouette_score(X_train_scaled,kmeans.labels_)
    silhouette_coefficients.append(score)
```

```python
silhouette_coefficients
```

```python
## plotting silhouette score
plt.plot(range(2,11),silhouette_coefficients)
plt.xticks(range(2,11))
plt.xlabel("Number of Cluters or k-value")
plt.ylabel("Silhoutte Coeffecient")
plt.show()
```

## fom the above curve
## when k=3 the line reaches its value more towards +1
## so according to silhoute the more the value towards +1 the more good the model is
## therefor 3- clusters are enough

## ------------------------------------------------------------------------------------------------------------------------

## therefor according to silhoute and knee locator we can came to conclusion that
##   3 clusters is enough to separate the data

```python
kmeans=KMeans(n_clusters=3,init="k-means++")
```

```python
kmeans.fit_predict(X_train_scaled)
y_pred=kmeans.predict(X_test_scaled)
```

```python
plt.scatter(X_test[:,0],X_test[:,1],c=y_pred)
```
