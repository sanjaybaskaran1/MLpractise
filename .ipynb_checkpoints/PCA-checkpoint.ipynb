```python
df = pd.DataFrame(data=cancer_dataset.data, columns=cancer_dataset.feature_names)
```

```python
df.head()
```

## in the above dataset all the features are my independednt features since this is pca algorithm i dont need to worry about dependent feature because im just going to reducte the dimensions of the independent features. 

## -- it is necessry to scale the data before feeding it into the pca model

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
pd.DataFrame(scaled_data)
```

## let me check total no.of features

```python
len(df.columns)
```

## with the help of pca lets reduce the (dimension 30) to (dimension-2)
## which basically means we're extracting only 2 features from those 30 features
## pc-1, pc-2 (pc - principle component)
## obviously the pc1 will have (more variance) fro the data than pc2

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
pca
```

```python
pca.fit(scaled_data)
```

```python
pca_data = pca.transform(scaled_data)
```

```python
pca_data
```

```python
pd.DataFrame(pca_data)
```

## so now we successfully extracted only 2-features from 30-features
## using pca
## lets see the variance of those 2 features

```python
pca.explained_variance_
```

## by using this 2-features we can able to plot and see our data distribution
## therefore with the help of pca we can reduce the feature and we plot it and visuually see the distribution of data

```python
plt.figure(figsize=(8,6))
plt.scatter(x=pca_data[:,0], y=pca_data[:,1], c=cancer_dataset.target, cmap='plasma' )
plt.xlabel('first-principle-component')
plt.ylabel('second-principle-component')
```
