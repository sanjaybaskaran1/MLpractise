## algerian forest fire prediction using linear regression models
## MultipleLinearRegression, LassoRegression, RidgeRegression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
path = r"C:\Users\ASUS\krish ml\Complete-Data-Science-With-Machine-Learning-And-NLP-2024\3-Complete Linear Regression\Practicals\Algerian_forest_fires_cleaned_dataset.csv"
```

```python
df = pd.read_csv(path)
```

```python
df.head()
```

```python
df.drop(columns={'day','month','year'}, inplace=True)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
df['Classes'] = np.where(df['Classes'].str.contains('not fire'),0,1)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
df.info()
```

```python
x = df.drop('FWI', axis=1)
```

```python
x.head()
```

```python
y = df['FWI']
```

```python
y.head()
```

```python
x.corr()
```

```python
corr_matrix = x.corr()
corr_matrix.iloc[1,0]
```

```python
corr_matrix.columns[0]
```

```python
def correlation_finder(df, threshold):
    high_corr_columns = set()
    corr_matrix = x.corr()
    for i in range(0, len(corr_matrix.columns)):
        for j in range(0, i):
            if abs(corr_matrix.iloc[i,j] > threshold):
                column_name = corr_matrix.columns[i]
                high_corr_columns.add(column_name)
    return high_corr_columns
            

multicolinearity = correlation_finder(df=x, threshold=0.85)
print(multicolinearity)
```

```python
x.head()
```

```python
print(type(x),x.shape,type(y),y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
print(type(x_train),x_train.shape,type(y_train),y_train.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
print(type(x_train_scaled),x_train_scaled.shape)
```

```python
x_train_scaled
```

```python
len(x_train.columns)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr
```

```python
lr.fit(x_train_scaled,y_train)
```

```python
y_pred_train = lr.predict(x_train_scaled)
```

```python
y_pred_train
```

```python
print(type(y_pred_train),y_pred_train.shape)
```

```python
plt.scatter(y_train, y_pred_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mae = mean_absolute_error(y_train, y_pred_train)
mse = mean_squared_error(y_train,y_pred_train)
r2 = r2_score(y_train,y_pred_train)
print("train data performance metrics:",mae,mse,r2)
```

## test data model checking

```python
y_pred_test = lr.predict(x_test_scaled)
```

```python
y_pred_test
```

```python
plt.scatter(y_test, y_pred_test)
```

```python
mae = mean_absolute_error(y_test, y_pred_test)
mse = mean_squared_error(y_test,y_pred_test)
r2 = r2_score(y_test,y_pred_test)
print("test data performance metrics:",mae,mse,r2)
```

## assumptions

## LassoCV 

```python
from sklearn.linear_model import LassoCV
```

```python
lasso = LassoCV(cv=5)
lasso
```

```python
lasso.fit(x_train_scaled,y_train)
```

```python
y_test_pred2 = lasso.predict(x_test_scaled)
y_test_pred2
```

```python
lasso.coef_
```

```python
mae = mean_absolute_error(y_test, y_test_pred2)
mse = mean_squared_error(y_test,y_test_pred2)
r2 = r2_score(y_test,y_test_pred2)
print("test data performance metrics:",mae,mse,r2)
```

## RidgeCV

```python
from sklearn.linear_model import RidgeCV
```

```python
ridge = RidgeCV(cv=5)
ridge
```

```python
ridge.fit(x_train_scaled,y_train)
```

```python
y_pred_test3 = ridge.predict(x_test_scaled)
```

```python
y_pred_test3
```

```python
residuals3 = y_test - y_pred_test3
```

```python
residuals3
```

```python
mae = mean_absolute_error(y_test, y_pred_test3)
mse = mean_squared_error(y_test,y_pred_test3)
r2 = r2_score(y_test,y_pred_test3)
print("test data performance metrics:",mae,mse,r2)
```
