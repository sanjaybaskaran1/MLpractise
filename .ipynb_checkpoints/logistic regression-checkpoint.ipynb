```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## import an library to ignore warnings

```python
import warnings
from sklearn.exceptions import ConvergenceWarning

# Suppress all warnings
warnings.filterwarnings("ignore")

# Or suppress specific warnings, like ConvergenceWarning
# warnings.filterwarnings("ignore", category=ConvergenceWarning)
```

```python
from sklearn.datasets import make_classification
```

```python
x,y = make_classification(n_samples=1000, n_features=10, n_classes=2, random_state=42)
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
m1 = LogisticRegression()
m1
```

```python
m1.fit(x_train, y_train)
```

```python
y_pred_test_m1 = m1.predict(x_test)
```

```python
y_pred_test_m1
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, classification_report, confusion_matrix
```

```python
accuracy_m1 = accuracy_score(y_pred_test_m1, y_test)
precision_m1 = precision_score(y_pred_test_m1, y_test)
recall_m1 = recall_score(y_pred_test_m1, y_test)
confu_matrix_m1 = confusion_matrix(y_pred_test_m1, y_test)
report_m1 = classification_report(y_pred_test_m1, y_test)

print(f'''accuracy_score:{accuracy_m1},
          precision_score:{precision_m1}
          recall_score:{recall_m1}
          confusion_matrix:
 {confu_matrix_m1}''')

print("--------------------------------------------")
print("CLASSIFICATION REPORT:")
print(report_m1)
```

## hyper parameter tuning and cv

## 1) GridSearchCV

```python
m2 = LogisticRegression()
m2
```

```python
penalty = ['l1', 'l2', 'elasticnet']
c_values =[100,10,1.0,0.1,0.01]
solver = ['lbfgs', 'liblinear', 'sag', 'saga']
```

```python
params = dict(penalty=penalty,
              C=c_values,
              solver=solver)
params
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
cv = StratifiedKFold()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(estimator=m2, param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1)
grid
```

```python
grid.fit(x_train,y_train)
y_pred_grid = grid.predict(x_test)
```

```python
y_pred_grid
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
accuracy_m2 = accuracy_score(y_pred_grid, y_test)
precision_m2 = precision_score(y_pred_grid, y_test)
recall_m2 = recall_score(y_pred_grid, y_test)
confu_matrix_m2 = confusion_matrix(y_pred_grid, y_test)
report_m2 = classification_report(y_pred_grid, y_test)

print(f'''accuracy_score:{accuracy_m2},
          precision_score:{precision_m2}
          recall_score:{recall_m2}
          confusion_matrix:
 {confu_matrix_m2}''')

print("--------------------------------------------")
print("CLASSIFICATION REPORT:")
print(report_m2)
```

## RandomizedSearchCV

```python
m3 = LogisticRegression()
m3
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
params
```

```python
random = RandomizedSearchCV(estimator=m3,param_distributions=params, scoring='accuracy', n_jobs=-1)
random
```

```python
random.fit(x_train,y_train)
```

```python
y_pred_m3 = random.predict(x_test)
```

```python
y_pred_m3
```

```python
random.best_params_
```

```python
random.best_score_
```

```python
accuracy_m3 = accuracy_score(y_pred_m3, y_test)
precision_m3 = precision_score(y_pred_m3, y_test)
recall_m3 = recall_score(y_pred_m3, y_test)
confu_matrix_m3 = confusion_matrix(y_pred_m3, y_test)
report_m3 = classification_report(y_pred_m3, y_test)

print(f'''accuracy_score:{accuracy_m3},
          precision_score:{precision_m3}
          recall_score:{recall_m3}
          confusion_matrix:
 {confu_matrix_m3}''')

print("--------------------------------------------")
print("CLASSIFICATION REPORT FOR M3:")
print(report_m3)
```
