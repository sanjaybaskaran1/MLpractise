```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000, n_features=2, n_classes=2,
                            n_clusters_per_class=2,n_redundant=0)
```

```python
X,type(X),X.shape
```

```python
y,type(y), y.shape
```

```python
pd.DataFrame(X)
```

```python
pd.DataFrame(X)[0] # --> feature1
```

```python
pd.DataFrame(X)[1] #-->feature-2 
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1], hue=y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
m1 = SVC()
m1
```

```python
m1.fit(x_train,y_train)
```

```python
y_pred_m1 = m1.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
m1_report = classification_report(y_test,y_pred_m1)
print("m1_report:","\n\n",m1_report)
print("\n")
m1_c_matrix = confusion_matrix(y_test,y_pred_m1)
print("m1_c_matrix:","\n\n",m1_c_matrix)
```

```python
param_grid = {
    "C":[0,0.1,1,10,100,1000],
    "gamma":[0,0.1,0.01,0.001,0.0001],
    "kernel":["liner","poly","rbf","sigmoid"]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(estimator=m1, param_grid=param_grid, refit=True, cv=5, verbose=3)
```

```python
grid.fit(x_train, y_train)
```

```python
grid.best_params_
```

```python
y_pred_grid_cv = grid.predict(x_test)
```

```python
y_pred_grid_cv_report = classification_report(y_test,y_pred_grid_cv)
print("y_pred_grid_cv_report:","\n\n",y_pred_grid_cv_report)
print("\n")
y_pred_grid_cv_c_matrix = confusion_matrix(y_test,y_pred_grid_cv)
print("y_pred_grid_cv_c_matrix:","\n\n",y_pred_grid_cv_c_matrix)
```
