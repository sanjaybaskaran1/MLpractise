```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
iris = load_iris()
```

```python
iris
```

```python
x = pd.DataFrame(data=iris['data'], columns=['sepal length In CM','sepal width In CM','petal length In CM','petal width In CM' ])
```

```python
x.head()
```

```python
y = iris.target
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.20, random_state = 10)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
m1 = DecisionTreeClassifier()
m1
```

```python
m1.fit(x_train,y_train)
```

```python
y_pred = m1.predict(x_test)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(m1,filled=True)
```

```python
m2 = DecisionTreeClassifier(max_depth=3)
m2
```

```python
m2.fit(x_train,y_train)
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(m2, filled=True)
```

```python
y_pred_m2 = m2.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
classification_report_m2 = classification_report(y_test, y_pred_m2)
print(classification_report_m2)
confusion_matrix_m2 = confusion_matrix(y_test, y_pred_m2)
print(confusion_matrix_m2)
```

## hyperparameter tunning

```python
params = {

    'criterion':['gini','entropy', 'log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2']

}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
decision_tree = DecisionTreeClassifier()
```

```python
grid = GridSearchCV(estimator=decision_tree, param_grid=params, cv=5, scoring='accuracy')
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred_grid = grid.predict(x_test)
```

```python
classification_report_grid = classification_report(y_pred_grid, y_test)
print(classification_report_grid)
```
