```python
import pandas as pd
import numpy as np
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r"C:\Users\ASUS\machne learning\Random forest\cleaned_data.csv")
```

```python
df.head()
```

```python
x = df.drop(columns=['ProdTaken'], axis=1)
```

```python
x.head()
```

```python
y = df['ProdTaken']
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20, random_state=42)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
numerical_features = [feature for feature in x if (x[feature].dtype != 'O')]
categorical_features = [feature for feature in x if feature not in numerical_features]
print("Total numerical features are:", len(numerical_features))
print("numerical features of x are: ", numerical_features, "\n")
print("Total categorical features are:", len(categorical_features))
print("categorical features of x are:","\n" ,categorical_features)
```

```python
numeric_transformer = StandardScaler()
categorical_transformer = OneHotEncoder()

preprocessor = ColumnTransformer(
                                   [
                                    ("OneHotEncoder", categorical_transformer, categorical_features),
                                    ("StandardScaler", numeric_transformer, numerical_features)
                                   ]
                                )

preprocessor
```

```python
x_train = preprocessor.fit_transform(x_train)
x_test = preprocessor.transform(x_test)
```

```python
x_train
```

```python
pd.DataFrame(x_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score, classification_report,ConfusionMatrixDisplay, \
                            precision_score, recall_score, f1_score, roc_auc_score,roc_curve 
```

```python
from xgboost import XGBClassifier
```

```python
models = {
    "Random Forest":RandomForestClassifier(),
    "Gradient Boost":GradientBoostingClassifier(),
    "Adaboost":AdaBoostClassifier(),
    "Xgboost":XGBClassifier()
}
```

```python
for i in range(0, len(models.keys())):
    ##initialize the model
    model = list(models.values())[i]

    ##train the model
    model.fit(x_train,y_train)

    ##make predictions
    y_pred_train = model.predict(x_train)
    y_pred_test = model.predict(x_test)

    ##model evaluation for(training_data) using performance metrics
    model_accuracy_train = accuracy_score(y_train, y_pred_train)
    model_precision_train = precision_score(y_train, y_pred_train)
    model_recall_train = recall_score(y_train, y_pred_train)
    model_f1_train = f1_score(y_train, y_pred_train)

    ##model evaluation for(test_data) using performance metrics
    model_accuracy_test = accuracy_score(y_test, y_pred_test)
    model_precision_test = precision_score(y_test, y_pred_test)
    model_recall_test = recall_score(y_test, y_pred_test)
    model_f1_test = f1_score(y_test, y_pred_test)

    ##print the model name
    print(list(models.keys())[i])

    ##print model performance for training set 
    print('Model performance for Training set')
    print("- ACCURACY {:.2f}".format(model_accuracy_train))
    print("- F1-SCORE {:.2F}".format(model_f1_train))
    print("- PRECISSION {:.2f}".format(model_precision_train))
    print("- RECALL {:.2f}".format(model_recall_train))


    print('-'*35)


    ##print model performance for testing set 
    print('Model performance for testing set')
    print("- ACCURACY {:.2f}".format(model_accuracy_test))
    print("- F1-SCORE {:.2F}".format(model_f1_test))
    print("- PRECISSION {:.2f}".format(model_precision_test))
    print("- RECALL {:.2f}".format(model_recall_test))

    print('='*35)
    print('\n')
```

```python
## Hyperparameter Training
rf_params = {"max_depth": [5, 8, 15, None, 10],
             "max_features": [5, 7, "auto", 8],
             "min_samples_split": [2, 8, 15, 20],
             "n_estimators": [100, 200, 500, 1000]}

adaboost_param={"n_estimators":[50,60,70,80,90],
                 "algorithm":['SAMME','SAMME.R']}

gradient_params={"loss": ['log_loss','deviance','exponential'],
                 "criterion": ['friedman_mse','squared_error','mse'],
                 "min_samples_split": [2, 8, 15, 20],
                 "n_estimators": [100, 200, 500],
                  "max_depth": [5, 8, 15, None, 10]}

xgboost_params = {"learning_rate": [0.1, 0.01],
                  "max_depth": [5, 8, 12, 20, 30],
                  "n_estimators": [100, 200, 300],
                  "colsample_bytree": [0.5, 0.8, 1, 0.3, 0.4]}
```

```python
# Models list for Hyperparameter tuning
randomcv_models = [
                   ("Xgboost", XGBClassifier(), xgboost_params)
]
```

```python
from sklearn.model_selection import RandomizedSearchCV

model_param = {}
for name, model, params in randomcv_models:
    random = RandomizedSearchCV(estimator=model,
                                   param_distributions=params,
                                   n_iter=100,
                                   cv=3,
                                   verbose=2,
                                   n_jobs=-1)
    random.fit(x_train, y_train)
    model_param[name] = random.best_params_

for model_name in model_param:
    print(f"---------------- Best Params for {model_name} -------------------")
    print(model_param[model_name])
```

```python
model = XGBClassifier(n_estimators=300, max_depth=12, learning_rate=0.1, colsample_bytree=0.8)

##train the model
model.fit(x_train,y_train)

##make predictions
y_pred_train = model.predict(x_train)
y_pred_test = model.predict(x_test)

##model evaluation for(training_data) using performance metrics
model_accuracy_train = accuracy_score(y_train, y_pred_train)
model_precision_train = precision_score(y_train, y_pred_train)
model_recall_train = recall_score(y_train, y_pred_train)
model_f1_train = f1_score(y_train, y_pred_train)

##model evaluation for(test_data) using performance metrics
model_accuracy_test = accuracy_score(y_test, y_pred_test)
model_precision_test = precision_score(y_test, y_pred_test)
model_recall_test = recall_score(y_test, y_pred_test)
model_f1_test = f1_score(y_test, y_pred_test)

##print the model name
print(list(models.keys())[i])

##print model performance for training set 
print('Model performance for Training set')
print("- ACCURACY {:.2f}".format(model_accuracy_train))
print("- F1-SCORE {:.2F}".format(model_f1_train))
print("- PRECISSION {:.2f}".format(model_precision_train))
print("- RECALL {:.2f}".format(model_recall_train))


print('-'*35)


##print model performance for testing set 
print('Model performance for testing set')
print("- ACCURACY {:.2f}".format(model_accuracy_test))
print("- F1-SCORE {:.2F}".format(model_f1_test))
print("- PRECISSION {:.2f}".format(model_precision_test))
print("- RECALL {:.2f}".format(model_recall_test))

print('='*35)
print('\n')
```
